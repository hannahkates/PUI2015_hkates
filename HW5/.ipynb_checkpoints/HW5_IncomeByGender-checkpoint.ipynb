### Importing data and creating subsets by gender

```python
url='http://cosmo.nyu.edu/~fb55/UI_CUSP_2015/data/'
allmales = {}
allfemales = {}

allmales['all'] = pd.read_excel(url+'/pinc08_2_1_1.xls',header=9)
allmales['white'] = pd.read_excel(url+'/pinc08_2_1_4.xls', header=9)
allmales['black'] = pd.read_excel(url+'/pinc08_2_1_6.xls', header=9)
allmales['hispanic'] = pd.read_excel(url+'/pinc08_2_1_9.xls', header=9)
allmales['asian'] = pd.read_excel(url+'/pinc08_2_1_8.xls', header=9)

allfemales['all']=pd.read_excel(url+'/pinc08_3_1_1.xls', header=9)
allfemales['white']=pd.read_excel(url+'/pinc08_3_1_4.xls', header=9)
allfemales['black']=pd.read_excel(url+'/pinc08_3_1_6.xls', header=9)
allfemales['hispanic']=pd.read_excel(url+'/pinc08_3_1_9.xls', header=9)
allfemales['asian']=pd.read_excel(url+'/pinc08_3_1_8.xls', header=9)
```

###Data cleanup

```python
for k in allmales:
    nonreal  = allmales[k].applymap(np.isreal)
    for c in allmales[k].columns :
        if nonreal.sum()[c]>1 and nonreal.sum()[c]<49:
            allmales[k][c][~nonreal[c]] = np.float('NaN')
            allmales[k][c] = allmales[k][c].astype(float)
```

```python
for k in allfemales:
    nonreal  = allfemales[k].applymap(np.isreal)
    for c in allfemales[k].columns :
        if nonreal.sum()[c]>1 and nonreal.sum()[c]<49:
            allfemales[k][c][~nonreal[c]] = np.float('NaN')
            allfemales[k][c] = allfemales[k][c].astype(float)
```

###Linear regression

####Numpy option

```python
idx = np.isfinite(allmales['all']['Median income'])&np.isfinite(allfemales['all']['Median income'])
poly = np.polyfit(allmales['all']['Median income'][idx], 
                  allfemales['all']['Median income'][idx], 1)
print poly

p = np.poly1d(poly)

pl.xlabel('Male Median Income from All Sources')
pl.ylabel('Female Median Income from All Sources')
pl.title('Fitting a line to male and female median income from all source types \n')
pl.plot(allmales['all']['Median income'][idx], p(allmales['all']['Median income'][idx]), color='blue')
pl.plot(allmales['all']['Median income'][idx], allfemales['all']['Median income'][idx], '.', color='black')
```

####Analytical method

```python
cor = stats.spearmanr(allmales['all']['Median income'][idx], 
                      allfemales['all']['Median income'][idx])
# m_all.std()/f_all.std()
beta1 = cor[0]*allfemales['all']['Median income'][idx].std()/allmales['all']['Median income'][idx].std() 
beta0 = allfemales['all']['Median income'][idx].mean()-beta1*allmales['all']['Median income'][idx].mean()

print beta1
print beta0

pl.plot(allmales['all']['Median income'][idx], beta1*allmales['all']['Median income'][idx] + beta0)
```

####Plot comparison

```python
pl.xlabel('Male Median Income from All Sources')
pl.ylabel('Female Median Income from All Sources')
pl.title('Fitting a line to male and female median income from all source types \n')
pl.plot(allmales['all']['Median income'][idx], p(allmales['all']['Median income'][idx]), color='blue', label='Numpy result')
pl.plot(allmales['all']['Median income'][idx], allfemales['all']['Median income'][idx], '.', color='black', label='Observed')
pl.plot(allmales['all']['Median income'][idx], beta1*allmales['all']['Median income'][idx] + beta0, color='red', label='Analytical result')
pl.legend()
```
